# Fetching electricity prices from ENTSO-E script

The purpose of this notebook was to initially test the ENTSO API pipeline and download our resplonse variable: the electricity price across different bidding zones. We also tested the format in which data was returned. The results were then exported as CSV. 


This notebook can be used by a successor group of the Marbl Energy project to quicly understand how to use ENTSO-E API

### Setting up the API pipeline

```python
#Setting up the API Key
TOKEN = "API-KEY" #enter your sectret key
```

```python
#Downloading the libraries
from entsoe import EntsoePandasClient
import pandas as pd
```

```python
#Setting up the client
client = EntsoePandasClient(api_key=TOKEN)
```

```python
#setting the time interval up
start_2015 = pd.Timestamp('20150101', tz='Europe/Brussels')    
end_2026 = pd.Timestamp('20260101', tz='Europe/Brussels')
```

```python
country_code = 'DE_LU' 
```

```python
germany_data = client.query_day_ahead_prices(country_code, start=start_2015, end=end_2026)
```

```python
#separating the output into 3 columns
df = germany_data.to_frame(name="price")
df["date"] = df.index.date                    
df["time"] = df.index.strftime("%H:%M")      
df.head(5)
```

```python
#saving the data on desktop 
df.to_csv(r"C:\Users\User\OneDrive\Radna površina\Data Science Lab\Data\DE_LU_Data.csv")
```
